Heart disease, encompassing a range of conditions affecting the heart, is a leading cause of death globally. It includes coronary artery disease, heart failure, and more. Risk factors include genetics, lifestyle choices, smoking, high blood pressure, diabetes, and high cholesterol. Early detection and prevention are crucial, as they can significantly improve outcomes.

Machine learning models can play a vital role in predicting heart disease by analyzing large datasets to identify patterns and risk factors. These models can use patient data-such as age, gender, medical history to predict the likelihood of heart disease.

# Analysis of chest pain type and its correlation with target variable

0: Typical angina (chest pain related to heart ischemia)

1: Atypical angina (chest pain not related to heart ischemia)

2: Non-anginal pain (not related to the heart)

3: Asymptomatic


```python
sns.countplot(data=data, y='cp')
plt.yticks([0, 1, 2, 3], ["Typical angina","Atypical angina","Non-anginal pain","Asymptomatic"])
plt.show()
```

```python
sns.countplot(x='cp', hue='target', data=data) 
plt.legend(labels=["No disease", "Disease present"]) 
plt.xticks([0, 1, 2, 3], ["Typical angina","Atypical angina","Non-anginal pain","Asymptomatic"], rotation=30)
plt.show()
```

# Analysis of fasting blood sugar and its correlation with target variable

# Filter the variables into categorical and continuous, and plot the continuous ones

```python
data.columns
```

```python
cate_val=[]
cont_val=[]

for i in data.columns:
    if data[i].nunique() <= 10:
        cate_val.append(i)
    else:
        cont_val.append(i)
print("categorical attributes are", cate_val)
print("continuous attributes are", cont_val)
```

```python
data.hist(cont_val)
plt.tight_layout()
plt.show()
# right or left skewness can be observed
```
